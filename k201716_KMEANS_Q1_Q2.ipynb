```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pp
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
data.columns=['cl1','cl2','cl3','cl4','label']
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
pd.DataFrame(data_scaled)
pd.DataFrame(data_scaled).describe()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++')
kmeans.fit(data_scaled)
```

```python
kmeans.inertia_
```

```python
SSE = []   
for cluster in range(1,20):
    kmeans = KMeans(n_clusters = cluster, init='k-means++')
    kmeans.fit(data_scaled)
    SSE.append(kmeans.inertia_)

frame = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})
pp.figure(figsize=(12,6))
pp.plot(frame['Cluster'], frame['SSE'], marker='o')
pp.xlabel('Number of clusters')
pp.ylabel('Inertia')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
dataset = pd.read_csv("D:\FastSemesters\semester5\ML\lab\emaan\ClusteringDataset.csv")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(dataset)

data=pd.DataFrame(data_scaled)
data.describe()
```

```python
SSE = []
for cluster in range(1,20):
  kmeans = KMeans(n_clusters = cluster, init='k-means++')
  kmeans.fit(data_scaled)
  SSE.append(kmeans.inertia_)
frame = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```

```python
kmeans = KMeans(n_clusters = 3, init='k-means++') 
kmeans.fit(data)
pred = kmeans.predict(data)
frame = pd.DataFrame(data)
frame['cluster'] = pred
frame['cluster'].value_counts()
```
