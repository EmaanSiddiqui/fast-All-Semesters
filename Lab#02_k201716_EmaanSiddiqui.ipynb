```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# Load CIFAR-10 dataset task 1
(cifar_train_images, cifar_train_labels), (cifar_test_images, cifar_test_labels) = tf.keras.datasets.cifar10.load_data()
print("cifar Train Images Shape:", cifar_train_images.shape)
print("cifar Train Labels Shape:", cifar_train_labels.shape)
print("cifar Test Images Shape:", cifar_test_images.shape)
print("cifar Test Labels Shape:", cifar_test_labels.shape)
```

```python
plt.figure(figsize=(10, 3)) #task 8
for i in range(5):
  plt.subplot(1,5,i+1)
  plt.imshow(cifar_train_images[i], cmap = 'gray')
  plt.title(f"label: {cifar_train_labels[i]}")
  plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
#task 9
train_images, train_labels = cifar_train_images[:50000], cifar_train_labels[:50000]
val_images, val_labels = cifar_train_images[50000:], cifar_train_labels[50000:]
test_images, test_labels = cifar_test_images, cifar_test_labels

# Display the shapes of the split datasets
print("Train Images Shape:", train_images.shape)
print("Train Labels Shape:", train_labels.shape)
print("Validation Images Shape:", val_images.shape)
print("Validation Labels Shape:", val_labels.shape)
print("Test Images Shape:", test_images.shape)
print("Test Labels Shape:", test_labels.shape)
```

```python
#task 11
import numpy as np
# Calculate class frequencies
class_counts = np.bincount(cifar_train_labels[:,0])
# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(cifar_train_labels, bins=range(11), rwidth=0.8, align='left', density=True)
plt.xticks(range(10))
plt.xlabel("Class Label")
plt.ylabel("Frequency")
plt.title("cifar Class Distribution Histogram")
plt.show()
# Plot bar plot
plt.figure(figsize=(8, 5))
plt.bar(range(10), class_counts, tick_label=range(10))
plt.xlabel("Class Label")
plt.ylabel("Frequency")
plt.title("cifar Class Distribution Bar Plot")
plt.show()
```
