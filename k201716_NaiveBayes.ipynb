# TASK #01

```python
def NB(df, test):
    
    #Calulating probability of both yes and no from evade
    total_count = df['EVADE'].count()
    outcome_yes = df[df['EVADE'] == 'Yes']
    total_yes = outcome_yes['EVADE'].count()
    outcome_no = df[df['EVADE'] == 'No']
    total_no = outcome_no['EVADE'].count()
    totalProbability_yes = total_yes / total_count
    totalProbability_no = total_no / total_count    
 
    #MARITAL STATUS
    MarSt_yes = 0
    MarSt_no = 0
    for i in outcome_yes['MARITAL STATUS']:
        if i == test[1]:
            MarSt_yes += 1
    for i in outcome_no['MARITAL STATUS']:
        if i == test[1]:
            MarSt_no += 1
    MarSt_yes = MarSt_yes /total_yes
    MarSt_no = MarSt_no / total_no
    
    #REFUND
    refund_yes = 0
    refund_no = 0
    for i in outcome_yes['REFUND']:
        if i == test[0]:
            refund_yes += 1
    for i in outcome_no['REFUND']:
        if i == test[0]:
            refund_no += 1
    refund_yes = refund_yes / total_yes
    refund_no = refund_no / total_no
    
    #TAXABLE INCOME
    mean_yes = outcome_yes['TAXABLE INCOME'].mean()
    mean_no = outcome_no['TAXABLE INCOME'].mean()
    variance_yes = outcome_yes['TAXABLE INCOME'].var()
    variance_no = outcome_no['TAXABLE INCOME'].var()
    taxationIncome_yes = (1 / (math.sqrt(2*math.pi*variance_yes))) * (math.exp(-((test[2] - mean_yes)**2) / (2*variance_yes)))
    taxationIncome_no = (1 / (math.sqrt(2*math.pi*variance_no))) * (math.exp(-((test[2] - mean_no)**2) / (2*variance_no)))
    
    probalityOf_yes = refund_yes * MarSt_yes * taxationIncome_yes * totalProbability_yes
    probabilityOf_no = refund_no * MarSt_no * taxationIncome_no * totalProbability_no
    
    if probalityOf_yes > probabilityOf_no:
        return 'Yes'
    else:
        return 'No'
    
```

```python
data = pd.DataFrame(dataframe)
data = data.set_index(data['TID'])
data = data.drop(['TID'], axis = 1)
print(data)
```

```python
data["REFUND"] = data["REFUND"].replace({"Yes":1,"No":0})
data["MARITAL STATUS"] = data["MARITAL STATUS"].replace({"Single":0,"Married":1,"Divorced":2})
data.at[4,"MARITAL STATUS"] = 1
data["EVADE"] = data["EVADE"].replace({"Yes":1,"No":0})
```

```python
prediction = NB(data, ['Yes', 'Divorced', 75])
print("Prediction Of Evasion: ", prediction)
```
