```python
import random
space = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z',' ']
Target = "gekko world this is ai kab"
class GeneticAlgorithm:
  def __init__(self):
    self.sample = []
    self.children =[[] for a in range(20)]
    self.generation = 0
    for a in range(0,20):
      self.sample.append([random.choice(space) for x in range(len(Target))])


  def crossover(self,parent):
    for a in range(0,19,2):
      chosen_parent = random.sample(parent,2)
      ch1,ch2=self.crossedchildren(chosen_parent)
      self.children[a]= ch1
      self.children[a+1]=ch2
    self.mutation()
  
  def mutation(self):
    for a in range(len(self.children)):
      for b in range(len(self.children[a])):
        selectgene = random.choice(range(len(self.children[a])))
        if self.children[a][selectgene] != Target[selectgene]:
          self.children[a][selectgene] = Target[selectgene]
          break
    self.generation+=1
    print("generation :", self.generation)
    a = self.best_child(self.children)
    for i in range(len(a)):
        print("".join(a[i]))
    bestchildren = "".join(a[0])
    print("best children of generation: ",bestchildren)
    self.sample = self.children
    if bestchildren != Target:
      self.selection()
  

  def selection(self):
    choicelist= []
    parent = [[] for a in range(5)]
    for a in range(0,20):
      val = self.fitnesscalculator(self.sample[a])
      choicelist.append([a,val])
    choicelist = sorted(choicelist,key=lambda x : x[1],reverse=True) #ascending in tsl
    for a in range(0,5):
      parent[a] = self.sample[choicelist[a][0]]
    self.crossover(parent)


  def fitnesscalculator(self,sentence):
    value = 0
    for a in range(len(sentence)):
      if sentence[a] == Target[a]:
        value+=10
      elif sentence[a] in Target[a]:
        value+=2
    return value
    
  def crossedchildren(self,parents):  #chromosomes are exchanging
    sliceposition = random.choice(range(int(len(Target)/2),len(Target)-1))
    temp1 = parents[0][:sliceposition] + parents[1][sliceposition:]
    temp2 = parents[1][:sliceposition] + parents[0][sliceposition:] 
    return temp1,temp2

  def best_child(self,childrenlist):
    choicelist=[]
    sortedchildren = []
    for a in range(0,20):
      val = self.fitnesscalculator(childrenlist[a])
      choicelist.append([a,val])
    choicelist = sorted(choicelist,key=lambda x : x[1],reverse=True)
    for a in range(0,20):
      sortedchildren.append(childrenlist[choicelist[a][0]])

    return sortedchildren

obj = GeneticAlgorithm()
obj.selection()

```
